```python
import tkinter as tk

window = tk.Tk()

window.title("Window")

var = tk.IntVar()
checkbutton = tk.Checkbutton(window, text="Chọn", variable=var)

checkbutton.pack()

label = tk.Label(window, text="Trạng thái: Không chọn")

label.pack()

def on_change_checkbutton():
    if var.get() == 1:
        label.config(text="Trạng thái: Đã chọn")
    else:
        label.config(text="Trạng thái: Không chọn")

checkbutton.config(command=on_change_checkbutton)

window.mainloop()
```
