```python
import threading
import random
import math

partial_maxes = []
lock = threading.Lock()

def calculate_partial_max(data_chunk, thread_id):
    if not data_chunk:
        return

    local_max = max(data_chunk)
    
    with lock:
        partial_maxes.append(local_max)
    
    data_str = "; ".join(map(str, data_chunk))
    print(f"Thread {thread_id}: {data_str}; -> Max cục bộ: {local_max}")


def bai_6_10_tim_max_da_luong():
    global partial_maxes

    try:
        n = int(input("Nhập số phần tử (N): ")) 
        num_threads = int(input("Nhập vào số thread: ")) 
    except ValueError:
        print("Lỗi: Đầu vào phải là số nguyên.")
        return

    main_list = [random.randint(0, 100) for _ in range(n)]
    
    print(f"\nList ({n} phần tử): {main_list}")
    
    list_length = len(main_list)
    chunk_size = math.ceil(list_length / num_threads) 
    
    threads = []
    partial_maxes = [] 

    for i in range(num_threads):
        start_index = i * chunk_size
        end_index = min((i + 1) * chunk_size, list_length)
        
        if start_index >= list_length:
            break
            
        data_chunk = main_list[start_index:end_index]
        
        thread = threading.Thread(
            target=calculate_partial_max, 
            args=(data_chunk, i + 1), 
            name=f"MaxThread_{i + 1}"
        )
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
    if partial_maxes:
        final_max = max(partial_maxes)
    else:
        final_max = "Không tìm thấy"
    expected_max = max(main_list) if main_list else "Không tìm thấy"
    print(f"Kiểm tra bằng hàm max() chuẩn :{expected_max}")
    
    print(f"\nMax cục bộ tìm được bởi các Threads: {partial_maxes}")
    print(f"Giá trị lớn nhất trong list (Tìm bằng đa luồng) = {final_max}")
bai_6_10_tim_max_da_luong()
```
