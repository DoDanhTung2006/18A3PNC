```python
import threading
import math
from functools import reduce

partial_results = [] 
lock = threading.Lock()

def calculate_partial_product(start, end):
    current_product = 1
    for i in range(start, end + 1):
        current_product *= i
    
    with lock:
        partial_results.append(current_product)
    
    print(f"[{threading.current_thread().name}] Tính tích từ {start} đến {end}. Kết quả cục bộ: {current_product}")

def main(n, num_threads=4):
    global partial_results
    partial_results = []
    
    print(f"TÍNH GIAI THỪA CỦA {n} BẰNG {num_threads} LUỒNG")
    
    threads = []
    chunk_size = math.ceil(n / num_threads) 

    for i in range(num_threads):
        start = i * chunk_size + 1
        end = min((i + 1) * chunk_size, n)

        if start > n:
            break
        
        thread = threading.Thread(target=calculate_partial_product, args=(start, end), name=f"Factorial_Thread_{i+1}")
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
    final_result = reduce(lambda x, y: x * y, partial_results)
    math_result = math.factorial(n)
    
    print(f"\nKết quả đa luồng ({len(partial_results)} tích): {final_result}")
    print(f"Kết quả thư viện math.factorial({n}): {math_result}")
    print(f"Kết quả tính toán khớp: {final_result == math_result}")

main(15, num_threads=4)
```
