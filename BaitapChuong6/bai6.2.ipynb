```python
import threading
import time
import random

def download_file(file_name):
    thread_name = threading.current_thread().name
    download_time = random.uniform(1, 3) 
    
    print(f"[{thread_name}] Bắt đầu tải xuống: {file_name}. Thời gian dự kiến: {download_time:.2f}s")
    time.sleep(download_time)
    print(f"[{thread_name}] ✅ Hoàn thành tải xuống: {file_name}.")

def main():
    files_to_download = ["File_A.zip", "Image_B.jpg", "Document_C.pdf", "Video_D.mp4"]
    threads = []

    print(f"MÔ PHỎNG TẢI XUỐNG ĐỒNG THỜI {len(files_to_download)} TỆP")

    start_time = time.time()

    for file in files_to_download:
        thread = threading.Thread(target=download_file, args=(file,))
        threads.append(thread)
        thread.start()
        
    for thread in threads:
        thread.join()

    end_time = time.time()
    print(f"\nTổng thời gian hoàn thành: {end_time - start_time:.2f} giây.")

main()
```
