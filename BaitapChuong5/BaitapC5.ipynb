### Bài 5.10

```python
import sqlite3

# Kết nối CSDL
conn = sqlite3.connect("product.db")
cur = conn.cursor()

# Tạo bảng product
cur.execute("""
CREATE TABLE IF NOT EXISTS product (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL NOT NULL,
    amount INTEGER NOT NULL
)
""")
conn.commit()


# ===== CÁC CHỨC NĂNG =====

def show_products():
    cur.execute("SELECT * FROM product")
    rows = cur.fetchall()

    if len(rows) == 0:
        print("Danh sách sản phẩm trống!")
    else:
        print("Danh sách sản phẩm:")
        for row in rows:
            print(row)


def add_product():
    name = input("Nhập tên sản phẩm: ")
    price = float(input("Nhập giá: "))
    amount = int(input("Nhập số lượng: "))

    cur.execute(
        "INSERT INTO product (name, price, amount) VALUES (?, ?, ?)",
        (name, price, amount)
    )
    conn.commit()
    print("Đã thêm sản phẩm!")


def search_product():
    name = input("Nhập tên sản phẩm cần tìm: ")

    cur.execute(
        "SELECT * FROM product WHERE name LIKE ?",
        ('%' + name + '%',)
    )
    rows = cur.fetchall()

    if len(rows) == 0:
        print("Không tìm thấy sản phẩm!")
    else:
        print("Kết quả tìm kiếm:")
        for row in rows:
            print(row)


def update_product():
    pid = int(input("Nhập id sản phẩm cần cập nhật: "))
    price = float(input("Nhập giá mới: "))
    amount = int(input("Nhập số lượng mới: "))

    cur.execute(
        "UPDATE product SET price = ?, amount = ? WHERE id = ?",
        (price, amount, pid)
    )
    conn.commit()

    if cur.rowcount == 0:
        print("Không tìm thấy sản phẩm!")
    else:
        print("Cập nhật thành công!")


def delete_product():
    pid = int(input("Nhập id sản phẩm cần xóa: "))

    cur.execute("DELETE FROM product WHERE id = ?", (pid,))
    conn.commit()

    if cur.rowcount == 0:
        print("Không tìm thấy sản phẩm!")
    else:
        print("Đã xóa sản phẩm!")


# ===== MENU CHÍNH =====

while True:
    print("\n----- QUẢN LÝ SẢN PHẨM -----")
    print("1. Hiển thị danh sách sản phẩm")
    print("2. Thêm sản phẩm")
    print("3. Tìm kiếm sản phẩm theo tên")
    print("4. Cập nhật giá và số lượng theo id")
    print("5. Xóa sản phẩm theo id")
    print("0. Thoát")

    choice = input("Chọn chức năng: ")

    if choice == "1":
        show_products()
    elif choice == "2":
        add_product()
    elif choice == "3":
        search_product()
    elif choice == "4":
        update_product()
    elif choice == "5":
        delete_product()
    elif choice == "0":
        print("Thoát chương trình!")
        break
    else:
        print("Lựa chọn không hợp lệ!")

# Đóng kết nối
conn.close()
```

```python
import sqlite3

# Kết nối CSDL (chưa có thì tự tạo)
conn = sqlite3.connect("ql_nhan_vien.db")
cursor = conn.cursor()
cursor.execute("DROP TABLE IF EXISTS NHAN_VIEN")
cursor.execute("DROP TABLE IF EXISTS PHONG")
conn.commit()

# Tạo bảng PHONG
cursor.execute("""
CREATE TABLE IF NOT EXISTS PHONG (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL NOT NULL,
    amount INTEGER NOT NULL
)
""")

# Tạo bảng NHAN_VIEN
cursor.execute("""
CREATE TABLE IF NOT EXISTS NHAN_VIEN (
    id INTEGER PRIMARY KEY,
    ho_ten TEXT NOT NULL,
    tuoi INTEGER NOT NULL,
    dia_chi TEXT NOT NULL,
    luong REAL NOT NULL,
    id_phong INTEGER,
    FOREIGN KEY (id_phong) REFERENCES PHONG(id)
)
""")

# Thêm dữ liệu vào PHONG
cursor.execute("""
INSERT INTO PHONG VALUES
(1, 'Phòng Kế toán', 5000000, 10),
(2, 'Phòng Nhân sự', 4500000, 8),
(3, 'Phòng Kỹ thuật', 6000000, 15)
""")

# Thêm dữ liệu vào NHAN_VIEN
cursor.execute("""
INSERT INTO NHAN_VIEN VALUES
(1, 'Nguyễn Văn A', 30, 'Hà Nội', 12000000, 1),
(2, 'Trần Thị B', 28, 'Hải Phòng', 10000000, 2),
(3, 'Lê Văn C', 35, 'Đà Nẵng', 15000000, 3),
(4, 'Phạm Thị D', 26, 'Hà Nội', 9000000, 1)
""")

conn.commit()
print("Tạo CSDL và dữ liệu thành công!")

conn.close()
```
