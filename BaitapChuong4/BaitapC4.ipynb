```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
# 1. Từ ser1 hãy xóa các mục có mặt trong ser2
filtered_ser1 = ser1[~ser1.isin(ser2)]
print(filtered_ser1)

# 2. Lấy tất cả các mục của ser1 và ser2 nhưng không nằm chung trong cả 2
unique_values = pd.Series(list(set(ser1).symmetric_difference(set(ser2))))
print(unique_values)
```

### Bài 4.5


```python
import numpy as np
import pandas as pd

ser = pd.Series(np.random.normal(10, 5, 25))

# 1. Giá trị nhỏ nhất (minimum)
min_val = ser.min()

# 2. Phần centile thứ 25 (25th percentile)
p25 = ser.quantile(0.25)

# 3. Trung vị (median)
median_val = ser.median()

# 4. Phần centile thứ 75 (75th percentile)
p75 = ser.quantile(0.75)

# 5. Giá trị lớn nhất (maximum)
max_val = ser.max()

print("Min:", min_val)
print("25th percentile:", p25)
print("Median:", median_val)
print("75th percentile:", p75)
print("Max:", max_val)
```

## Bài 4.7

```python
import numpy as np
import pandas as pd

np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

# 1. Tìm 2 giá trị có tần suất cao nhất
top2 = ser.value_counts().nlargest(2).index

# 2. Thay phần tử không thuộc top 2 = 'Other'
result = ser.where(ser.isin(top2), 'Other')

print("Series ban đầu:")
print(ser)
print("\nSeries sau khi xử lý:")
print(result)
```

## Bài 4.8

```python
import numpy as np
import pandas as pd

ser = pd.Series(np.random.random(20))

# 1. Tạo nhãn phân vị từ D1 đến D10
labels = [f"D{i}" for i in range(1, 11)]

# 2. Dùng qcut để chia thành 10 phân vị
ser_deciled = pd.qcut(ser, 10, labels=labels)

print("Series ban đầu:")
print(ser)

print("\nSeries sau khi thay bằng phân vị:")
print(ser_deciled)
```

```python
import pandas as pd

# Đọc file euro12.csv (đặt file cùng thư mục với file .py hoặc notebook)
euro12 = pd.read_csv(r"C:\Users\Dell N5430\OneDrive\Đính kèm\PNC\18A1PNC\chuong_04\Euro2012.csv")

# . In type, shape, danh sách cột
print("== THÔNG TIN EURO12 ==")
print(type(euro12))
print(euro12.shape)
print(euro12.columns)

# 1. In giá trị cột Goals
print("\n== CỘT GOALS ==")
print(euro12["Goals"])

# 2. Có bao nhiêu đội tham gia Euro2012?
print("\n== SỐ ĐỘI BÓNG ==")
print("Số đội:", euro12["Team"].nunique())

# 3. In thông tin của Euro2012
print("\n== INFO ==")
print(euro12.info())

# 4. Tạo data frame discipline
discipline = euro12[["Team", "Yellow Cards", "Red Cards"]]
print("\n== DISCIPLINE ==")
print(discipline)

# 5. Sắp xếp giảm dần theo Red Cards, Yellow Cards
discipline_sorted = discipline.sort_values(
    by=["Red Cards", "Yellow Cards"], ascending=False
)
print("\n== DISCIPLINE SẮP XẾP ==")
print(discipline_sorted)

# 6a. Tính trung bình Yellow Cards
print("\n== TRUNG BÌNH YELLOW CARDS ==")
print(discipline["Yellow Cards"].mean())

# 6b. Lọc đội ghi hơn 6 bàn
print("\n== ĐỘI GHI > 6 BÀN ==")
print(euro12[euro12["Goals"] > 6])

# 7. In các đội tên bắt đầu bằng chữ G
print("\n== ĐỘI BẮT ĐẦU BẰNG 'G' ==")
print(euro12[euro12["Team"].str.startswith("G")])

# 8. In 7 cột đầu
print("\n== 7 CỘT ĐẦU ==")
print(euro12.iloc[:, :7])

# 9. In tất cả cột trừ 3 cột cuối
print("\n== TRỪ 3 CỘT CUỐI ==")
print(euro12.iloc[:, :-3])

# 10. In các cột Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print("\n== 5 CỘT QUAN TRỌNG ==")
cols = ["Team", "Goals", "Shooting Accuracy", "Yellow Cards", "Red Cards"]
print(euro12[cols])

# 11. Chỉ hiển thị Shooting Accuracy của England, Italy, Russia
print("\n== SHOOTING ACCURACY - 3 ĐỘI ==")
print(euro12[euro12["Team"].isin(["England", "Italy", "Russia"])]
      [["Team", "Shooting Accuracy"]])
```

## Bài 4.21

```python
import pandas as pd

drink = pd.read_csv("drinks.csv")

print("=== Cột ===")
print(drink.columns)

print("\n=== 5 dòng đầu ===")
print(drink.head())

print("\n=== 5 dòng cuối ===")
print(drink.tail())

# Bia trung bình theo châu lục
print("\n=== BIA TRUNG BÌNH THEO CHÂU LỤC ===")
print(drink.groupby("continent")["beer_servings"].mean())

# Thống kê rượu vang theo châu lục
print("\n=== THỐNG KÊ RƯỢU VANG ===")
print(drink.groupby("continent")["wine_servings"].describe())

# Trung bình các loại bia + rượu theo châu lục
print("\n=== TRUNG BÌNH BIA + RƯỢU THEO CHÂU LỤC ===")
print(drink.groupby("continent")[["beer_servings","spirit_servings","wine_servings"]].mean())

# Median
print("\n=== MEDIAN THEO CHÂU LỤC ===")
print(drink.groupby("continent")[["beer_servings","spirit_servings","wine_servings"]].median())

# Spirit min/max/mean
print("\n=== SPIRIT: MEAN - MIN - MAX ===")
print(drink.groupby("continent")["spirit_servings"].agg(["mean","min","max"]))

# Sort tăng dần theo bia
print("\n=== SORT BIA TĂNG DẦN ===")
print(drink.sort_values("beer_servings").head())

# 5 nước uống bia nhiều nhất
print("\n=== TOP 5 UỐNG BIA NHIỀU NHẤT ===")
print(drink.sort_values("beer_servings", ascending=False).head())

# 5 nước uống bia ít nhất
print("\n=== TOP 5 UỐNG BIA ÍT NHẤT ===")
print(drink.sort_values("beer_servings").head())
```
